```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# pandas correlation dataframe 
data = pd.DataFrame({
    'physics':[10,9,8,7],
    'chemistry':[5,7,6,9],
    'mathematics':[10,15,12,15],
    'Bio': [12,13,14,15]
})
data.corr()
```

```python
# data aggregation
data.agg('mean')
```

```python
# data visualization
data.plot()
```

```python
data.plot.bar()
```

```python
data.plot.hist()
data
```
