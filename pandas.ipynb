# Data Normalization with pandas

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame([
    [15000,1000,100,100000],
    [10000,9,3000,7000],
    [500,1500,1000,2000],
    [1300,1400,1500,1600]
], columns = [ 'colA', 'colB','colC','colD'])
print(data)
data.plot(kind= 'bar')
plt.show()
print(data.describe())
data.plot(kind = 'hist')
plt.show()
```

```python
## Maximum Absolute value Scaling
df = data.copy()
for column in df.columns:
    df[column]= df[column]/df[column].abs().max()
print(df)
df.plot(kind = 'bar')
plt.show()
```

```python
print(df.describe())
```

```python
# Min-Max Scaling 
df1= df.copy()
for column in df1.columns:
    df1[column] = (df1[column]-df1[column].min())/(df1[column].max()-df1[column].min())
print(df1)
    
```

```python
df2 = (df -df.min())/(df.max() - df.min())
print(df2)
```

```python
# Z- score
df3 = data.copy()
std= df3.std()
mean = df3.mean()
z = (data - mean)/std
print(z)
```
